```python
# Operators
# Arithmetic Operators
#+ --------> normal addition
# - --------> subtraction
# * --------> multiplication
# / --------> division(without decimal part)
# // --------> floor division(decimal part is included)
# % --------> modulo(gives the remainder)
# **  ------> exponentiation

# Assignment Operators
# = --------> assigns a value to a variable
# += --------> adds a value to the variable and assigns the result to the variable
# example: x += 10 is equivalent to x = x + 10
# -= --------> subtracts a value from the variable and assigns the result to the variable
# example: x -= 5 is equivalent to x = x - 5
# *= --------> multiplies the variable by a value and assigns the result to the variable
# example: x *= 2 is equivalent to x = x * 2
# /= --------> divides the variable by a value and assigns the result to the variable

# Comparison Operators
# == --------> equal to
# != --------> not equal to
# > ---------> greater than
# < ---------> less than
# >= --------> greater than or equal to
# <= --------> less than or equal to

# Logical Operators
# AND --------> returns True if both statements are true
# OR ---------> returns True if at least one statement is true
# NOT --------> returns True if the statement is false
```

```python
a = 10
b = 5
c = a > b
print(c)

P="Parth"
K="Krishna"
N=P<K
print(N)
```

```python
# Nested if-else
x = 150

if x > 100:
    print("Value is greater than 100")
    
    if x % 2 == 0:
        print("even")
    else:
        print("odd")

else:
    print("Value is less than or equal to 100")  
```
