```python
#zip()
#function is used to combine two or more iterables (like lists, tuples, etc.) into a single iterable of tuples. Each tuple contains elements from the corresponding positions of the input iterables. The resulting iterable will have a length equal to the shortest input iterable.
list_A=[x for x in range(1,9)]
list_B=[A,B,C,D,E,F,G,H]
NEW_d=zip(list_A,list_B)
print(list(NEW_d))
```
